```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
pkmn_type_colors = ['#78C850',  # japan
                    '#F08030',  # india
                    '#6890F0',  # china
                    '#A8B820',  # thailand
                    '#A8A878',  # vietnam
                    '#A040A0',  # cambodia
                    '#F8D030',  # singapore
                    '#E0C068',  # indonesia
                    '#EE99AC',  # canada
                    '#C03028',  # united_states
                    '#F85888',  # united_kingdom
                    '#B8A038',  # germany
                    '#705898',  # france
                    '#98D8D8',  # spain
                    '#7038F8',  # philippines
                    '#5BC7E5',  # south_korea
                    '#1c9ca5',  # mexico
                    '#6840A0',  # poland
                    '#52206a',  # portugal
                    '#6A1748',  # egypt
                    '#5BC7E5',  # denmark
                    '#09CAFF',  # australia
                    '#5BC7E5',  # cambodia
                    '#e2ceae',  # belgium
                    '#5BC7E5',  # sudan
                    '#E55BAD',  # russia
                    '#dc0e88',  # norway
                    '#E5D05B',  # saudi_arabia     
                   ]
```

```python
sns.set(style="ticks")
plt.xticks(rotation=-40)  #x tick label rotation 
age_structure1= age_structure.sort_values(by='65+', ascending = False)  #sort values 
age_graph1= sns.barplot(x='country name',y='65+', data=age_structure1, palette=pkmn_type_colors) #draw bar plot 
age_graph1.figure.set_size_inches(20,9) #figure size 
age_graph1.axes.set_title('Aged Population by Country', fontsize=34,color="b",alpha=0.5)  #title
age_graph1.set_xlabel("Country Name",size = 20,color="r",alpha=0.5) #x_label setting
age_graph1.set_ylabel("Population % over 65",size = 20,color="g",alpha=0.5) #y_label setting
age_graph1.tick_params(labelsize=14,labelcolor="black")
```

```python
import seaborn as sns
from matplotlib.gridspec import GridSpec

sns.set(style="whitegrid")
#sns.set_color_codes("Spectral")

plt.figure(2, figsize=(33,24)) #figure size
the_grid = GridSpec(2, 2)  #subplots  

plt.subplot(the_grid[0, 1],  title='Population % over 65+')
sns.barplot(x='country name',y='65+', data=age_structure1, palette=pkmn_type_colors) #bar plot setting  
plt.xticks(rotation=-50) # x label tick rotation 
plt.subplot(the_grid[0, 0], title='%Age 25-54')
sns.barplot(x='country name',y='25-54', data=age_structure1, palette=pkmn_type_colors)
plt.xticks(rotation=-50)
plt.subplot(the_grid[1, 1],  title='%Age 15-24')
sns.barplot(x='country name',y='15-24', data=age_structure1, palette=pkmn_type_colors)
plt.xticks(rotation=-50)
plt.subplot(the_grid[1, 0], title='%Age 54-64')
sns.barplot(x='country name',y='54-64', data=age_structure1, palette=pkmn_type_colors)
plt.xticks(rotation=-50)

plt.suptitle('Age Structure', fontsize=30,color="b",alpha=0.5)  #title setting 
age_graph1.tick_params(labelsize=9,labelcolor="blue")
```
